# ðŸ’¡ Intuitive Approach to Extract Digits from a Number

To solve this problem, the intuitive approach is to:

## ðŸ§© Steps
- **Isolate the last digit** of the number using the modulus operation (`% 10`)
- **Print the isolated digit**
- **Remove the last digit** from the number using integer division (`// 10`)
- **Repeat the process** until all digits have been extracted

## ðŸ§  Why This Works
- The modulus operation (`% 10`) returns the **remainder** when a number is divided by 10, which is the **last digit** of the number.
- Integer division (`// 10`) removes the **last digit** from the number.

## ðŸ” Result
Repeating these steps will eventually extract and print **all digits in reverse order**.


```python
def extract_digit(n):
    digits = []
    while n > 0:
        temp = n % 10
        digits.append(temp)
        n = n // 10
    return digits

result = extract_digit(123)
print(result)
```
