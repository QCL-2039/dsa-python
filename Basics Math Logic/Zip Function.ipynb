Print name + mark only for students whose marks are above class average.

```python
names = ["A", "B", "C", "D"]
marks = [78, 85, 62, 90]

avg = sum(marks) / len(marks)

result = list(filter(lambda x: x[1] > avg, zip(names, marks)))
print(result)
```

Remove negative numbers

Square the remaining numbers

Store the result in a list

```python
nums = [3, -2, 5, -7, 8, -1]

result = list(map(lambda x: x**2, filter(lambda x: x > 0, nums)))
print(result)
```

Find how many unique classes exist

Print each class and its count

```python
labels = ["cat", "dog", "cat", "bird", "dog", "dog"]

unique = set(labels)
print("Number of classes:", len(unique))

for cls in unique:
    print(cls, labels.count(cls))
```

Sort features by importance (descending)

salary 0.4
age 0.25
education 0.2
experience 0.15


```python
features = ["age", "salary", "experience", "education"]
importance = [0.25, 0.4, 0.15, 0.2]

result = sorted(
    zip(importance, features),
    key=lambda x: x[0],
    reverse=True
)

for score, feature in result:
    print(feature, score)

```

Remove students with scores < 70

Square remaining scores

Sort students by squared score (descending)

Print index, name, score

```python
names = ["A", "B", "C", "D"]
scores = [70, 85, 60, 90]

students=sorted((map((lambda x:(x[0],x[1]**2)),filter(lambda x: x[1]>=70,zip(names,scores)))),key=lambda x:x[1],reverse=True)

for idx,(name,score) in enumerate(students):
    print(idx,name,score)
```

```python
names = ["A", "B", "C", "D"]
scores = [70, 85, 60, 90]

filtered = filter(lambda x: x[1] >= 70, zip(names, scores))
squared = map(lambda x: (x[0], x[1]**2), filtered)
sorted_result = sorted(squared, key=lambda x: x[1], reverse=True)

for i, (name, score) in enumerate(sorted_result):
    print(i, name, score)
```

```python
names = ["A", "B", "C", "D"]
scores = [70, 85, 60, 90]

students = sorted(
    map(lambda x: (x[0], x[1] ** 2),
        filter(lambda x: x[1] > 70, zip(names, scores))
    ),
    key=lambda x: x[1],
    reverse=True
)

for idx, (name, score) in enumerate(students):
    print(idx, name, score)
```
