```python
#1
prices = [100, 200, 300]
tax = [10, 20, 30]
final_price=list(map((lambda x:x[0]+x[1]),zip(prices,tax)))
final_price
```

```python
#2
names = ["A", "B", "C", "D"]
attendance = [80, 65, 90, 70]

filtered_students=list(filter(lambda x:x[1]>=75,zip(names,attendance)))
filtered_students
```

```python
#3
scores = [45, 60, 30, 80, 90]

pass_stud=list(filter(lambda x: x>=50,scores))
print(len(pass_stud))
```

```python
#4
cities = ["Dhaka", "Chittagong", "Dhaka", "Sylhet", "Sylhet"]
print("Unique cities are:",set(cities))
```

```python
#5
data = [(1, 10), (2, 5), (3, 8)]

sorted_data=sorted(data,key=lambda x:x[1],reverse=False)
sorted_data
```

```python
#6
values = [5, 10, 15, 20]

mn,mx=min(values),max(values)

norm_value=list(map((lambda x:(x-mn)/mx-mn),values))
norm_value
```

```python
#7
marks = [70, 85, 90, 75]

print("Average:",sum(marks)/len(marks))
```
