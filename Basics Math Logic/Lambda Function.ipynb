What map() does

map(function, iterable):

Applies the function to each element

Returns a map object (converted to list here)

```python
numbers=[i for i in range(2,12,2)]
print(numbers)

def squares(num):
    return num**2

result=list(map(squares,numbers))
result
```

```python
result=list(map(lambda x:x**2,numbers))
result
```

```python
result=list(filter(lambda x: x%2 == 0,numbers))
result
```

```python

from functools import reduce
nums=[i for i in range(1,11)]

result=reduce(lambda acc,x: acc+x,nums)
result
```

```python
import random
from functools import reduce

nums = [random.randint(10, 100) for i in range(10)]
nums

max_num=reduce(lambda acc,x: acc if acc >x else x,nums)
print(nums)
max_num
```
