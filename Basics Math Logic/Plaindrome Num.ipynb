```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        
        # Negative numbers are never palindromes
        if x < 0:
            return False

        original = x
        reversed_num = 0

        while x > 0:
            last_digit = x % 10
            reversed_num = reversed_num * 10 + last_digit
            x //= 10

        return reversed_num == original
```
