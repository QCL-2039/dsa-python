```python
def count_digit(n):
    n=abs(n)
    count=0
    
    if n==0:
        return 1
    else:
        while n>0:
            n=n//10
            count+=1
        return count
print(count_digit(0))        
```

# Optimal Solution with time and space complexity of O(1)

```python
import math
def count_digit(n):
    n=abs(n)
    
    if n==0:
        return 1
    else:
        return int(math.log10(n)+1)
print(count_digit(0))
```
