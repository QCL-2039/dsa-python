```python
def reverse_num(n):
    n=abs(n)
    reverse_num=0
    count=0
    x=n

    
    while n>0:
        n=n//10
        count+=1
    print(count)

    while count>0:
        temp=x%10
        x=x//10
        reverse_num+=temp*(10**(count-1))
        count-=1
    return reverse_num
print(reverse_num(-120))

        
```

```python
class Solution:
    def reverse(self, x) :
        is_negative = False
        if x < 0:
            is_negative = True
        num = abs(x)
        answer = 0
        while num > 0:
            last_digit = num % 10
            answer = (answer * 10) + last_digit
            num //= 10

        if answer < (-(21**31)) or answer > (2**31 - 1):
            return 0
        return -answer if is_negative else answer
    
```
